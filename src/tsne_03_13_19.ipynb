```python
# import statements
import pandas as pd
from sklearn.cluster import KMeans
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import sklearn.preprocessing as preproc
import seaborn as sb
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
```

```python
tfidf_path = '/Users/jackie16201/Desktop/Lab/clustering/data/tfidf'
tfidf = pd.read_csv(tfidf_path)
```

```python
tfidf_nolabel = tfidf.iloc[:,1:]
tfidf_scaled = preproc.scale(tfidf_nolabel)
```

```python
# utilize my test set
test_path = '/Users/jackie16201/Desktop/Lab/clustering/data/'
test_set = pd.read_csv(test_path + 'test_drugs', header=None)

# get the test set drugs from the tfidf matrix
test_set_strings = test_set.iloc[:,1]
tfidf_drug_names = tfidf.iloc[:,0]

# use boolean indexing to get the logical array of the intersection of the test set and tfidf
logical_drug_arr = tfidf_drug_names.isin(test_set_strings)
tfidf_test = tfidf_nolabel[logical_drug_arr]
```

```python
def tryNumComponents(numComp):
    pca = sklearn.decomposition.PCA(n_components=numComp)
    pca.fit(tfidf_scaled)                 
    print 'first n components: ' + str((pca.explained_variance_ratio_))
    print 'sum: ' + str((np.sum(pca.explained_variance_ratio_)))
```

```python
# need to PCA first so it's computationally tractable to do tSNE
# based on results above, looks like doing numComponents = 4000 works fine, ~90% of variance
pca_4000 = sklearn.decomposition.PCA(n_components=4000)
pca_4000.fit(tfidf_scaled)                 
```

```python
print 'first n components: ' + str((pca_4000.explained_variance_ratio_))
print 'sum: ' + str((np.sum(pca_4000.explained_variance_ratio_)))
```

```python
# transform scaled matrix
tfidf_pca_4000 = pca_4000.transform(tfidf_scaled)
```

```python
# try tSNE on this
tfidf_embed = TSNE().fit_transform(tfidf_pca_4000)
print tfidf_embed.shape
```

```python
plt.scatter(tfidf_embed[:,0], tfidf_embed[:,1])
```

```python
# kmeans cluster on this bad boy
kmeans = KMeans(n_clusters=2, random_state=0).fit(tfidf_embed)
kmeans_labels = kmeans.labels_.reshape(6100,1).flatten()

dim1 = tfidf_embed[:,0]
dim2 = tfidf_embed[:,1]


plt.figure(figsize=(8,10))
plt.scatter(dim1[kmeans_labels == 0], dim2[kmeans_labels == 0], color = 'blue')
plt.scatter(dim1[kmeans_labels == 1], dim2[kmeans_labels == 1], color = 'red')

plt.xlabel('tsne one')
plt.ylabel('tsne two')
plt.legend(('Cluster 1', 'Cluster 2'))
plt.savefig('2kmeans_cluster_tsne.png')
```

```python
def printUniqueDrugs(unique_drug_list):
    unique_drug_names = []
    for num_drug in range(0, len(unique_drug_list)):
        curr_drug = unique_drug_list.iloc[num_drug]
        split_words = curr_drug.split("-")
        drug_name = split_words[0]
        if drug_name not in unique_drug_names:
            unique_drug_names.append(drug_name)

    print unique_drug_names
```
